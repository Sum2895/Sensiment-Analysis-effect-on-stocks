```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Load the NLTK stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load the NLTK WordNet lemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
data = pd.read_csv('/content/drive/MyDrive/data mining/Combined_News_DJIA.csv')
```

```python
def preprocess_text(text):
    if isinstance(text, str):  # Check if the input is a string
        text = text.lstrip('b')
        text = text.lower()
        # Remove special characters from the text
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Tokenize the text into words
        words = word_tokenize(text)

        # Remove stop words from the text
        filtered_words = [word for word in words if word.casefold() not in stop_words]

        # Perform lemmatization on the text
        lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

        # Join the preprocessed words back into a single string
        preprocessed_text = ' '.join(lemmatized_words)

        return preprocessed_text
    else:
        return text  # Return the input unchanged if it is not a string

for col_name in data.iloc[:,2:]:
    data[col_name] = data[col_name].apply(preprocess_text)
```

```python
def get_subjectivity(text):
  return TextBlob(text).sentiment.subjectivity

def get_polarity(text):
  return TextBlob(text).sentiment.polarity
```

```python
data['subjectivity'] = data['combined_news'].apply(get_subjectivity)
data['polarity'] = data['combined_news'].apply(get_polarity)
```

```python
def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment= sia.polarity_scores(text)
    return sentiment
```

```python
compound=[]
neg=[]
pos=[]
neu=[]
SIA=0

for i in range (0, len(data['combined_news'])):
    SIA= getSIA(data['combined_news'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    pos.append(SIA['pos'])
    neu.append(SIA['neu'])
```

```python
data['compound']= compound
data['neg']= neg
data['pos']=pos
data['neu']=neu
```
